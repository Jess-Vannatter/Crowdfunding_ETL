```python
import pandas as pd
import numpy as np
```

```python
# Read the data into a Pandas Dataframe
crowdfunding_data = pd.ExcelFile("crowdfunding.xlsx")

# Get the sheet names 
crowdfunding_data.sheet_names
```

```python
# Get the crowdfunding_info from the crowdfunding_info worksheet
crowdfunding_info_df = pd.read_excel(crowdfunding_data, sheet_name='crowdfunding_info')
crowdfunding_info_df.head()
```

```python
# Checking end of crowdfunding_info_df
crowdfunding_info_df.tail()
```

```python
# Use .info to get more information on the crowdfunding_info_df - like the index, data types, non-null values amd memory usage 
crowdfunding_info_df.info()
```

```python
# Get the unique values in the catagory and subcatagory column
crowdfunding_info_df["category & sub-category"].unique()

## used Value_counts to find number of times these values were returned in the data 
# crowdfunding_info_df.drop_duplicates()["category & sub-category"].value_counts()

##.used .count to find number of unique catagary's/ sub-category's
# cat_subcat_unique_df = crowdfunding_info_df["category & sub-category"].drop_duplicates()
# cat_subcat_unique_df.count()
```

```python
# assign the category and subcategory values to category and subcategory columns 
crowdfunding_info_df[["category","subcategory"]] = crowdfunding_info_df["category & sub-category"].str.split('/', n=1, expand=True)
crowdfunding_info_df.head(10)
```

```python
# get the distinct values in the category and subcategory columns
print(crowdfunding_info_df["category"].nunique())
print(crowdfunding_info_df["subcategory"].nunique())
```

```python
# Get the unique categorys and subcategory's in to seperate lists 
categories = crowdfunding_info_df["category"].unique()

subcategories = crowdfunding_info_df["subcategory"].unique()

print(categories)

print()

print(subcategories)
```

```python
# Create Numpy arrays that 1-10 for the category_id's and 1-25 for the subcategory_ids
category_ids = np.arange(1, 10)
subcategory_ids = np.arange(1,25)

print(category_ids)
print(subcategory_ids)
```

```python
# use a list comprehension to add "cat0" " to each category _id
cat_ids = ["cat0" + str(cat_id) for cat_id in category_ids]

# use a list comprehension to add "sat0" " to each subcategory _id
scat_ids = ["scat0" + str(scat_id) for scat_id in subcategory_ids]

print(cat_ids)
print(scat_ids)
```

```python
# Create a category Dataframe with the cat_ids array as the category_id and categories lists as they category name
category_df = pd.DataFrame({
    "category_id" : cat_ids,
    "category" : categories})
```

```python
# Create a subcategory Dataframe with the scat_ids array as the subcategory_id and subcategories lists as they category name
subcategory_df = pd.DataFrame({
    "subcategory_id" : scat_ids,
    "subcategory" : subcategories})
```
